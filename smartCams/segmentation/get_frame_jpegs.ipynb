```python
import cPickle as pickle
from matplotlib import pyplot as plt
import os

all_images_path = '../input_output/all_images.pi'
# NOTE: output_dir must be absolute path
output_dir = '/Users/alvinyuan/Documents/Vision/smartCams/smartCams/segmentation/video_frame_jpegs/'
interesting_frames = [47, 49, 52, 55, 80]
```

```python
all_images = pickle.load(open(all_images_path, 'rb'))

print all_images.shape
```

```python
import cv2
# Save frames as jpegs
frame_filenames = []
for frame_num in interesting_frames:
    frame_filename = output_dir + "frame_" + str(frame_num) + ".jpg"
    frame_filenames.append(frame_filename)
    cv2.imwrite(frame_filename, all_images[frame_num])
```

```python
from selective_search_ijcv_with_python import selective_search
# Get windows
all_boxes = selective_search.get_windows(frame_filenames)
```

```python
# Show windows
# http://nbviewer.ipython.org/github/BVLC/caffe/blob/master/examples/detection.ipynb
im = plt.imread(frame_filenames[0])
plt.imshow(im)
currentAxis = plt.gca()

boxes = all_boxes[0]
for box in boxes:
    coords = (box[0], box[1]), box[2], box[3]
    currentAxis.add_patch(plt.Rectangle(*coords, fill=False, edgecolor='r', linewidth=1))

# det = df.iloc[j]
# coords = (det['xmin'], det['ymin']), det['xmax'] - det['xmin'], det['ymax'] - det['ymin']
# currentAxis.add_patch(plt.Rectangle(*coords, fill=False, edgecolor='b', linewidth=5))
```
